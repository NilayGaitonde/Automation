```python
import pandas as pd
import datetime
import PyPDF2
import numpy as np
```

```python
new_df = pd.read_csv('test.csv',index_col=0)
add_df = {
    "Products": ["3_PersonalLoan","2_BusinessLoan"],
    "Loan Institution":["OHMY Loan Private Limited","Bank123"],
    "date_opened": [datetime.date(2019, 1, 1),datetime.date(2019, 1, 2)],
    "Sanction/Credit Limit":[10000,12000],
    "Balance":[100000,200000],
    "EMI":[1000,2000],
    "Paid Principle": [21887,21993]
}
# new_df = new_df.append(add_df, ignore_index=True)
new_df = pd.concat([new_df,pd.DataFrame(add_df)])
new_df
```

```python
new_df.sort_values(by=['Paid Principle'],ascending=False,inplace=True)
new_df.shape
```

```python
top_up_df = new_df.loc[new_df['Products'].isin(['3_PersonalLoan','5_AutoLoan','6_HouseLoan'])]
# new_df.drop(top_up_df.index)
top_up_df = top_up_df.drop_duplicates(subset=['Loan Institution'],keep='first')
new_df = pd.concat([new_df,top_up_df])
new_df
```

```python
pivot = pd.pivot_table(new_df,values=['Paid Principle',"Balance"],index=['Products'],aggfunc=np.sum,fill_value=0)
top_up = pivot.loc["3_PersonalLoan","Paid Principle"]
# pivot = pivot.reset_index()
pivot = pivot.loc[pivot.index[0:np.where(pivot.index == "3_PersonalLoan")[0][0]]]
pivot
```

```python
top_up
```

```python
balance = pivot["Balance"].to_list()
# while top up is not 0 keep subtracting from the balance
outstandingBalance = top_up
for i in range(len(balance)):
    productBalance = balance[i]
    if(productBalance>top_up):
        print(f"Reduce {pivot.index[i]}")
        outstandingBalance = productBalance - top_up
        top_up = 0
        print(f"Remaining Balance: {outstandingBalance}")
    elif(top_up>productBalance):
        print(f"Remove {pivot.index[i]}")
        outstandingBalance = top_up - productBalance
        print(f"Remaining top up: {outstandingBalance}")
```
